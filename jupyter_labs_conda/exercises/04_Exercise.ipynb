# Exercise #4

Step 1. Using the following [cities table](https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population), produce 2 dictionaries using the top 10 cities.

- Dictionary 1: Key - City Name, Value - 2020 Estimate
- Dictionary 2: Key - City Name, Value - 2010 Actual

```python
# Step 1 Solution

dict1 = {
    "New York": 8253213,
    "Los Angeles": 3970219,
    "Chicago": 2677643,
    "Houston": 2316120,
    "Phoenix": 1708127,
    "Philadelphia": 1578487,
    "San Antonio": 1567118,
    "San Diego": 1422420,
    "Dallas": 1343266,
    "San Jose": 1013616,
}

dict2 = {
    "New York": 8175133,
    "Los Angeles": 3792621,
    "Chicago": 2695598,
    "Houston": 2100263,
    "Phoenix": 1445632,
    "Philadelphia": 1526006,
    "San Antonio": 1327407,
    "San Diego": 1307402,
    "Dallas": 1197816,
    "San Jose": 945942,
}

display(dict1)
display(dict2)
```

Step 2. Produce a data series for each dictionary: 2020 Estimate and 2010 Actual. Display both data series below.

```python
# Step 2 Solution

import pandas as pd

series1 = pd.Series(dict1)

display(series1)

series2 = pd.Series(dict2)

display(series2)
```

Step 3. Produce a data frame using the two data series. Display the data frame below.

```python
# Step 3 Solution

cities = pd.DataFrame({
  "pop2020": series1,
   "pop2010": series2,
})

cities
```

Step 4. Add two new columns to the data frame for the name of the state and the 2016 land area.

```python
# Step 4 Solution

cities["state"] = pd.Series({
    "New York": "NY",
    "Los Angeles": "CA",
    "Chicago": "IL",
    "Houston": "TX",
    "Phoenix": "AZ",
    "Philadelphia": "PA",
    "San Antonio": "TX",
    "San Diego": "CA",
    "Dallas": "TX",
    "San Jose": "CA",
})

cities["land area"] = pd.Series({
    "New York": 301.5,
    "Los Angeles": 468.7,
    "Chicago": 227.3,
    "Houston": 637.5,
    "Phoenix": 517.6,
    "Philadelphia": 134.2,
    "San Antonio": 461.0,
    "San Diego": 325.2,
    "Dallas": 340.9,
    "San Jose": 177.5,
})

cities
```

Step 5. Calculate a new column for the percent change in population from 2010 to 2020. Confirm your calculations with the numbers in Wikipedia.

```python
# Step 5 Solution

cities["pop change"] = ((cities["pop2020"] - cities["pop2010"]) / cities["pop2010"]) * 100

cities
```

Step 6. Calculate two columns for population density for both 2010 and 2020.

```python
# Step 6 Solution

cities["pop density 2020"] = cities["pop2020"] / cities["land area"]
cities["pop density 2010"] = cities["pop2010"] / cities["land area"]

cities
```

Step 7. Save the data frame to a CSV.

```python
# Step 7 Solution

cities.index.name = "city"

cities.to_csv("cities.csv")
```

Step 8. Swap the columns and indexes. Save to a CSV.

```python
# Step 8 Solution

cities_swapped = cities.stack().unstack(0)
cities_swapped.index.name = "Stats"

display(cities)
display(cities_swapped)


cities_swapped.to_csv("cities-swapped.csv")
```
